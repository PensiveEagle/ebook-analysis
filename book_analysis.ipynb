```python
pattern = re.compile("[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
findings = re.findall(pattern, book)
print(len(findings))
findings
```

## What are the most used words?

```python
pattern = re.compile("[a-zA-z]+")
findings = re.findall(pattern, book.lower())

words = {}
for word in findings:
    if word in words.keys():
        words[word] = words[word] + 1
    else:
        words[word] = 1

words_list = [(value, key) for (key, value) in words.items()]
sorted(words_list, reverse=True)[:10]
```

# Using Natural Language Processing

## Most used words non-articles

```python
words_with_articles = sorted(words_list, reverse=True)
```

```python
import nltk
```

```python
nltk.download( "stopwords" )
from nltk.corpus import stopwords
```

```python
english_stopwords = stopwords.words( "english" )
```

```python
words_without_articles = []
for count, word in words_with_articles:
    if word not in english_stopwords:
        words_without_articles.append( (count, word) )
```

## Sentiment Analysis - Most positive / most negative chapters

```python
nltk.download("vader_lexicon")
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyser = SentimentIntensityAnalyzer()
```

```python
pattern = re.compile( "Chapter [0-9]+" )
chapters = re.split( pattern, book )[1:]
scores_table = []
```

```python
chapter_number = 0
for index, chapter in enumerate( chapters ):
    scores = analyser.polarity_scores( chapter )
    scores["chapter"] = index + 1
    scores_table.append( scores )
```

```python
scores_table
```

```python
negative_score = 0
positive_score = 0
for row in scores_table:
    if row["neg"] > negative_score:
        negative_score = row["neg"]
        most_negative_chapter = f"Chapter {row["chapter"]}"
    if row["pos"] > positive_score:
        positive_score = row["neg"]
        most_positive_chapter = f"Chapter {row["chapter"]}"
```

```python
print( f"The most positive chapter is {most_positive_chapter} with a score of {positive_score}" )
print( f"The most negative chapter is {most_negative_chapter} with a score of {negative_score}" )
```
